10) Give me the solution to $x=\cos(x)$ that is at most $10^{-6}$ away from the real answer.

```python
def f(x):
    return x - math.cos(x)

def bisection(f, l, r,n):
    left = l
    right = r
    mid = (l+r)/2.0
    assert(f(left)*f(right) < 0)
    for i in range(n):
        mid = (left+right)/2.0
        if (f(mid)* f(left) < 0):
            right = mid
        else:
            left = mid
    return mid

bisection(f,0.2,1.2,19)
```
