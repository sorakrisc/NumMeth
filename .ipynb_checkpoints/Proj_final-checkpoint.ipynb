```python
outputImage = np.zeros((1456,2190, 4))
# print(InputImage[:,:,0])
np.mean(InputImage[:,:,0])
```

```python

meanRedChannel =np.mean(np.mean(InputImage[:,:,0]))

meanGreenChannel = np.mean(np.mean(InputImage[:,:,1]))
meanBlueChannel = np.mean(np.mean(InputImage[:,:,2]))
```

```python
L=0.5
LRed = meanRedChannel/L
Lblue = meanBlueChannel/L
LGreen = meanGreenChannel/L
print(InputImage[:,:,0]*(1/LRed))
plt.imshow(InputImage)
```

```python
L=0.5
LRed = meanRedChannel/L
Lblue = meanBlueChannel/L
LGreen = meanGreenChannel/L
print(InputImage[:,:,0]*(1/LRed))

outputImage[:,:,0] = InputImage[:,:,0]*(1/LRed)
outputImage[:,:,1] = InputImage[:,:,1]*(1/LGreen)
outputImage[:,:,2] = InputImage[:,:,2]*(1/Lblue)
outputImage[:,:,3] = InputImage[:,:,3]

for iy in range(0,ny):
    for ix in range(0,nx):
        img = outputImage[iy,ix][:-1]
        hsv=matplotlib.colors.rgb_to_hsv(img)
#         if (hsv[1]<0.2 or hsv[2]>0.9):
#             #it is light color
# #             print("H",end="")
#             hsv = np.array([0,0,hsv[2]-value])
#             img = matplotlib.colors.hsv_to_rgb(hsv)
#             outputImage[iy,ix] = np.array([img[0]*(1/LRed),
#                                            img[1]*(1/LGreen),
#                                            img[2]*(1/Lblue),
#                                            1.])
#         maxouthsv =matplotlib.colors.rgb_to_hsv(outputImage[iy,ix][:-1])
        maxouthsv =hsv
        if(max(maxouthsv)>1):
            maxouthsv = (maxouthsv/(max(maxouthsv)))*0.9
            outputImage[iy,ix][:-1] = matplotlib.colors.hsv_to_rgb(maxouthsv)

plt.figure()
plt.imshow(outputImage)
plt.figure()
plt.imshow(badimg)
# plt.imshow(outputImage[1100:1150,270:350])
# print(badimg[1100:1150,270:350][0][0])
# print(outputImage[1100:1150,270:350][0][0])
# print(goodimg[1100:1150,270:350][0][0])
plt.figure()
plt.imshow(goodimg)
```
